```python
class Node:
    def __init__(self,val):
        self.val = val
        self.next = None
        
class LinkedList:
    def __init__(self):
        self.head = None
        
    def insert(self,index,val):
        
        new_node = Node(val)
        
        #case 1 
        if index == 0:
            
            new_node.next = self.head
            self.head = new_node
            return
        
        counter = 0 
        temp = self.head
        while temp is not None and counter < index:
            
            prev = temp
            temp = temp.next
            
            counter += 1
            
        prev.next = new_node
        new_node.next = temp
        return
    
    def remove(self,val):
        
        temp = self.head
        
        #case 1
        if temp.val == val:
            self.head = temp.next
            return True
        
        #case2 
        while temp is not None:
            if temp.val == val:
                break
                
            temp = temp.next
            prev = temp
            
        if temp is None:
            #not Found case 2.1
            return None
        
        #found case 2.2
        prev.next = temp.next
        return True
        
    def __str__(self):
        ret = "["
        temp = self.head
        while temp is not None:
            ret += str(temp.val) + " , "
            temp = temp.next
  
        ret += "]"
        return ret
    
    def get_length(self):
        
        counter = 0
        temp = self.head
        while temp is not None:
            temp = temp.next
            counter += 1
            
        return counter
    
    def push(self,val):
            length = self.get_length()
            self.insert(length,val)
            return
        
    def get(self,index):
        
        if self.head is None:
            raise IndexError("No list is made yet")
        
        temp = self.head
        for i in range(index):
            value = temp.val
            temp = temp.next
            
        return value
    
    def pop(self):
        self.remove(self.get(self.get_length()))
        return 
```

```python
l = LinkedList()
l.insert(0,1)
l.insert(1,2)
l.insert(2,3)
l.push(4)

print(l)
l.pop()
print(l)

l.remove(1)
l.remove(2)
#l.remove(3)
print(l)
```
